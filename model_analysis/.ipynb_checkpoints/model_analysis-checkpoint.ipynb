## Imports

```python
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
from keras.layers import Dropout
from keras import regularizers
from keras import optimizers
import keras
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import operator
%matplotlib inline
```

## Create Test, Dev and Training set

```python
mapped_classes = {0: 'apple_pie', 1: 'beignets', 2: 'bruschetta', 3: 'chicken_wings', 4: 'chocolate_cake', 5: 'eggs_benedict', 6: 'fried_calamari', 7: 'fried_rice', 8: 'greek_salad', 9: 'guacamole', 10: 'hamburger', 11: 'miso_sup', 12: 'panna_cotta', 13: 'pizza', 14: 'pork_chop', 15: 'prime_rib', 16: 'ramen', 17: 'risotto', 18: 'sashimi'}
```

```python
batch_size = 32
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
```

```python
test_dev_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
training_set = train_datagen.flow_from_directory('../data/training/',
                                                 target_size = (64, 64),
                                                 batch_size = 32,
                                                 class_mode = 'categorical')
```

```python
dev_set = test_dev_datagen.flow_from_directory('../data/dev/',
                                            target_size = (64, 64),
                                            batch_size = 32,
                                            class_mode = 'categorical')
```

```python
test_set = test_dev_datagen.flow_from_directory('../data/test/',
                                            target_size = (64, 64),
                                            batch_size = 32,
                                            class_mode = 'categorical')
```

## Create Neural Network

### Simplest Model - No Regularization - No Dropout

```python
model = Sequential()
```

```python
model.add(Conv2D(32, (3, 3), input_shape = (64, 64, 3), activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Conv2D(32, (3, 3), activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Flatten())
model.add(Dense(units = 128, activation = 'relu'))
model.add(Dense(units = 19, activation = 'softmax'))
```

```python
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit_generator(training_set,
                         steps_per_epoch = 594,
                         epochs = 5,
                         validation_data = test_set,
                         validation_steps = 60)
```

```python
#Store the first model
model.save('simplest_model')
```

```python
model = keras.models.load_model('simplest_model')
```

#### Testing Out the Simplest Model with a Burger, Ramen and Wings

```python
#Loading Images
burger_image = image.load_img('../data/test/hamburger/215037.jpg', target_size = (64, 64))
ramen_image = image.load_img('../data/test/ramen/181732.jpg', target_size = (64, 64))
wing_image = image.load_img('../data/test/chicken_wings/274528.jpg', target_size = (64, 64))
burger_image = image.img_to_array(burger_image)
burger_image = np.expand_dims(burger_image, axis = 0)
ramen_image = image.img_to_array(ramen_image)
ramen_image = np.expand_dims(ramen_image, axis = 0)
wing_image = image.img_to_array(wing_image)
wing_image = np.expand_dims(wing_image, axis = 0)
```

```python
#Plotting images
plt.imshow(mpimg.imread('../data/test/hamburger/215037.jpg'))
```

```python
plt.imshow(mpimg.imread('../data/test/ramen/181732.jpg'))
```

```python
plt.imshow(mpimg.imread('../data/test/chicken_wings/274528.jpg'))
```

```python
def print_top5_prediction(prediction):
    prediction = prediction[0]
    dictionary_result = {}
    for i in range(prediction.shape[0]):
        if prediction[i] > 0.:
            dictionary_result[mapped_classes[i]] = prediction[i]
    dictionary_result = sorted(dictionary_result.items(), key=operator.itemgetter(1), reverse=True)
    print "-----Top 5 of Probabilities-----"
    for class_name, probabily in dictionary_result:
        print "Class: {0} : {1:.2f}%".format(class_name, probabily * 100)
    print "--------------------------------"
```

```python
prediction_hamburger = model.predict_proba(burger_image)
prediction_wings = model.predict_proba(wing_image)
prediction_ramen = model.predict_proba(ramen_image)
print_top5_prediction(prediction_hamburger)
print_top5_prediction(prediction_wings)
print_top5_prediction(prediction_ramen)
```

```python
#One more testing using an image from another distribution to make predictions, 
#in this case Pizza from Pizza House's Facebook Page.
pizza_image = burger_image = image.load_img('../data/distribution/pizza_1.jpg', target_size = (64, 64))
pizza_image = image.img_to_array(pizza_image)
pizza_image = np.expand_dims(pizza_image, axis = 0)
```

```python
#Plotting image
plt.imshow(mpimg.imread('../data/distribution/pizza_1.jpg'))
```

```python
prediction_pizza = model.predict_proba(pizza_image)
print_top5_prediction(prediction_pizza)
```

### Simplest Model Enhanced - One Extra Layer, L2 Regularization and Dropout

```python
model_enhanced = Sequential()
```

```python
model_enhanced.add(Conv2D(32, (3, 3), input_shape = (64, 64, 3), activation = 'relu'))
model_enhanced.add(MaxPooling2D(pool_size = (2, 2)))
model_enhanced.add(Conv2D(32, (3, 3), activation = 'relu'))
model_enhanced.add(MaxPooling2D(pool_size = (2, 2)))
model_enhanced.add(Flatten())
model_enhanced.add(Dense(units = 128, activation = 'relu', kernel_regularizer=regularizers.l2(0.01)))
model_enhanced.add(Dropout(0.2))
model_enhanced.add(Dense(units = 64, activation = 'relu', kernel_regularizer=regularizers.l2(0.01)))
model_enhanced.add(Dense(units = 19, activation = 'softmax'))
```

```python
model_enhanced.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
model_enhanced.fit_generator(training_set,
                         steps_per_epoch = 594,
                         epochs = 5,
                         validation_data = test_set,
                         validation_steps = 60)
```

### Model using Data Augmentation by 128x128

```python
training_set_128x128 = train_datagen.flow_from_directory('../data/training/',
                                                 target_size = (128, 128),
                                                 batch_size = 32,
                                                 class_mode = 'categorical')
dev_set_128x128 = test_dev_datagen.flow_from_directory('../data/dev/',
                                            target_size = (128, 128),
                                            batch_size = 32,
                                            class_mode = 'categorical')
```

```python
model_128x128 = Sequential()
```

```python
model_128x128.add(Conv2D(64, (3, 3), input_shape = (128, 128, 3), activation = 'relu'))
model_128x128.add(MaxPooling2D(pool_size = (2, 2)))
model_128x128.add(Conv2D(64, (3, 3), activation = 'relu'))
model_128x128.add(MaxPooling2D(pool_size = (2, 2)))
model_128x128.add(Conv2D(32, (3, 3), activation = 'relu'))
model_128x128.add(MaxPooling2D(pool_size = (2, 2)))
model_128x128.add(Flatten())
model_128x128.add(Dense(units = 128, activation = 'relu'))
model_128x128.add(Dense(units = 19, activation = 'softmax'))
```

```python
model_128x128.compile(optimizer = optimizers.Adam(decay=2), loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
model_128x128.fit_generator(training_set_128x128,
                         steps_per_epoch = 594,
                         epochs = 1,
                         validation_data = dev_set_128x128,
                         validation_steps = 60)
```

## Due to the poor results obtained before, a change of strategy is a must.

For this new model, the next changes will be implemented:
- Bigger and cleaner training set.
- Cleaned dev set.
- No more test set.
- Increase the data augmentation.
- Create a new Architecture.
- Using a Pre-Trained model (VGG-16)

### Data Augmentation

```python
clean_training_datagen = ImageDataGenerator(
                            rotation_range=40,
                            width_shift_range=0.2,
                            height_shift_range=0.2,
                            rescale=1./255,
                            shear_range=0.2,
                            zoom_range=0.2,
                            horizontal_flip=True,
                            fill_mode='nearest')
```

```python
clean_validation_datagen = ImageDataGenerator(rescale=1./255)
```

```python
clean_training_data = clean_training_datagen.flow_from_directory(
                            '../cleaned_data/training/', 
                            target_size=(150, 150),
                            batch_size=batch_size,
                            class_mode='categorical')
```

```python
clean_validation_data = clean_validation_datagen.flow_from_directory(
                            '../cleaned_data/test/', 
                            target_size=(150, 150),
                            batch_size=batch_size,
                            class_mode='categorical')
```

```python
clean_model_v1 = Sequential()
```

```python
#First Conv Layer
clean_model_v1.add(Conv2D(32, (3,3), input_shape=(150, 150, 3), activation='relu'))
clean_model_v1.add(MaxPooling2D(pool_size = (2, 2)))
#Second Conv Layer
clean_model_v1.add(Conv2D(32, (3,3), activation='relu'))
clean_model_v1.add(MaxPooling2D(pool_size = (2, 2)))
#Third Conv Layer
clean_model_v1.add(Conv2D(64, (3,3), activation='relu'))
clean_model_v1.add(MaxPooling2D(pool_size = (2, 2)))
#Flatten
clean_model_v1.add(Flatten())
#Fully Connected Layer
clean_model_v1.add(Dense(128, activation='relu'))
#Output Layer
clean_model_v1.add(Dense(len(mapped_classes), activation='softmax'))
```

```python
clean_model_v1.compile(optimizer='adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
clean_model_v1.fit_generator(
                            clean_training_data,
                            steps_per_epoch=17040/batch_size,
                            epochs=10,
                            validation_data = clean_validation_data,
                            validation_steps = 1889/batch_size,    
                            use_multiprocessing= True
                            )
```

```python
clean_model_v1.save('clean_model_v2.pkl')
```
